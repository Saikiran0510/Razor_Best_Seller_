```python
file=pd.read_excel('Sample_Longlist_Data.xlsx')
df=pd.DataFrame(file)
df.head(10)                         #printing head of the data
```

```python
df.info()             #information about the data
```

```python
#removing all chinnese companies taking account of only those who is not from china
df=df[df['businessaddress'].str[-2:]!='CN']
df.info()
```

```python
#after removing chineese companies we left over only with 602 sellers out of 1839 sellers
# as instructed Razor does not acquire Chinese sellers at this point, so you can use this data to exclude sellers in China from your analysis.
```

```python
#now removing all those tuples whose sellerratings are not provided
```

```python
df=df.dropna(how='any', subset=['sellerratings'])
df.info()
```

```python
#after looking at this data we left with only 559 sellers
```

```python
indexNames = df[ df['Max % of negative seller ratings - last 30 days'] >df['Max % of negative seller ratings - last 90 days']].index
df=df.drop(indexNames)
print(df)
```

```python
#I have done this as we eant best company so we would like company which is growing
#or in other words seller whose negetive ratings is decreasing.
```

```python
df=df.dropna(how='any', subset=['sellerproductcount'])
#Now dropping if sellerproductcount is also not given in the data as we want best seller
df=df[df['sellerproductcount'].str.split().str[-2].str.replace(',','').astype(int)>1000] #And a condition that sellerproductcount should be >1000 products
df.head(10)   #printing the head 10 values
```

```python
df.info()
```

```python
#Now we left with only 134 sellers i have assumed that seller should have sold more than 1000 products in the past
```

```python
#Now taking account of only those sellers whose more than 90% percent in positive ratings
# we can also account of new sellers who have lifetime ratings but 
# As we want best seller so I am beleiving on past experience of a seller as well.
df=df[df['sellerratings'].str.split().str[0].str.replace('%','').astype(int)>90]
df.head(10)
```

```python
df.info()    #Now we left with only 81 sellers
```

```python
#Creating temporeary dataFrames so as to change some in it and then to restore it
df2=df
df2['temp']=df2['sellerratings'].str.split('(').str[1]   # as temp column having number of rating and some garbage string
df['col']=df2['temp'].str.split().str[0]                 # creating an extra column having number of people whos have given ratings to the seller
df=df.drop('temp',axis='columns')                        #deleting that extra created column
df=df[df['col'].astype(int)>1000]                         #Assuming the Threshhold number of ratiings to be 1000 
df.head(10)
```

```python
df.info()
```

```python
#Now we left with only 9 Sellers 
# Amoung which we can check about the hero product rating of 1st and 2nd best product
#Now sorting the data based on hero product ratings
```

```python
#Now sorting the DataFrame based on 'Count of seller brands','Hero Product 1 #ratings','Hero Product 2 #ratings' in decending order and the top seller is the best
df=df.sort_values(by=['Count of seller brands','Hero Product 1 #ratings','Hero Product 2 #ratings'],ascending=False)
df.head(10)
```
